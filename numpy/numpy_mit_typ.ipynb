```python
import numpy as np
import matplotlib as plt
```

```python
def square(x : np.ndarray) -> np.ndarray:
    '''
        jedes element im eingabe array quadrieren
    '''
    return np.power(x, 2)

def leaky_relu(x : np.ndarray) -> np.ndarray:
    return np.maximum(0.2*x,x)
```

```python
x = np.array([-3,-2,-1,0,1,2,3], dtype = np.float32)
x
```

```python
y1 = square(x)
y1
```

```python
y2 = leaky_relu(x)
y2
```

```python
import matplotlib.pyplot as plt
x = np.arange(-3, 3, 0.1)
plt.plot(x, square(x), c='b')
plt.plot(x, leaky_relu(x), c='r')
plt.grid(True)
```
