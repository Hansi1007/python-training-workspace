# Machine Learning
## The Chain Rule

```python
slope = 2.0

heigth = lambda weigth : slope * weigth

x_weigth = [i for i in range(0, 10, 2)]
y_heigth = []
for x in x_weigth:
    y_heigth.append(heigth(x))
plt.plot(x_weigth, y_heigth)

y_size = []
for x in y_heigth:
    y_size.append(1/4*x)

print(y_size)    

```

```python

def outer_func(x):
    y = 2 * x
    return lambda y: 1/4 * y

for i in range(0,10,2):
    closure = outer_func(i)
    print(f"closure({i+5}) = {closure(i+5)}")
```
