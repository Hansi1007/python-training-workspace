# Rectified Linear Unit (RELU) Function

https://www.codesansar.com/deep-learning/common-activation-functions-python.htm

The rectified linear activation function (RELU) is a piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero.

RELU has become the default activation function for different neural networks because a model that uses RELU is easier to train and it often achieves better performance.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Rectified Linear Unit
def relu(x):
  temp = [max(0,value) for value in x]
  return np.array(temp, dtype=float)
# Derivative for RELU
def drelu(x):
  temp = [1 if value>0 else 0 for value in x]
  return np.array(temp, dtype=float)
```

```python
print(relu([1,1.5,0.001,-0.1,-0.02]))
print(drelu([1,1.5,0.001,-0.1,-0.02]))

print(max(0, -1 ))
print(max(0, 0 ))
print(max(0, 1.5 ))
```

```python
# Generating data to plot
x_data = np.linspace(-6,6,100)
y_data = relu(x_data)
dy_data = drelu(x_data)
```

```python
# Plotting
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('RELU & Derivative')
plt.legend(['f(x)','f\'(x)'])
plt.grid()
plt.show()
```
