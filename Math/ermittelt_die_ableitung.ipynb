```python
from typing import Callable
import numpy as np
```

```python
def derive(func: Callable[[np.ndarray], np.ndarray], input_: np.ndarray, delta: float = 0.001) -> np.ndarray:
    '''
    Ermittelt die Ableitung einer Funktion "func" fÃ¼r jedes Element im Array "input_".
    '''
    return (func(input_ + delta) - func(input_ - delta)) / (2 * delta)

def square(x: np.ndarray) -> np.ndarray:
    return np.power(x,2)

def pow3(x):
    return np.power(x, 3)
```

```python
x = np.arange(-4,4,0.1)
yd = derive(square, x)
```

```python
import matplotlib.pylab as plt


#plt.figure(figsize=(4, 3))
plt.plot(x, square(x), c='r')
plt.plot(x, yd)
plt.grid(True)
```

```python

plt.figure(figsize=(4, 3))
plt.plot(x, pow3(x), c='r')
yd = derive(pow3, x)
plt.plot(x, yd)
plt.grid(True)
plt.show()
```
