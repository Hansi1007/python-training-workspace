```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import sympy as sy
```

```python

def f1(x):
    return 3*x

def f2(x):    
    return -2*x

```

```python
x = np.array([-3,-2,-1,0,1,2,3])
y1 = np.zeros(x.shape)

y1 = f1(x)
y1
```

```python
y2 = f2(f1(x))
```

```python
plt.plot(x, y1, 'r-')
plt.plot(x, y2, 'b-')
```

```python
print("Änderung von 1 an f1(1) = ", f1(1))
print("Änderung von 3 an f2(3) = ", f2(3))

print("Änderung von 1 an f2(f1(1)) = ", f2(f1(1)))
```
