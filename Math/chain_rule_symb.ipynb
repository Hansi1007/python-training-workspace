```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
sqare = lambda x: np.power(x,2)
leaky_Relu = lambda x: np.where(x > 0, x, x * 0.01)                          
```

```python
x,y,a,b,c = sy.symbols('x y a b c ')

f1 = x**2 + 2 *x 
f2 = sy.sin(f1)

sy.diff(f2,x)
```

```python
def f2f1(x):
    # f1 = x**2 f2 = sigmoid(x)  f2f1 = f2(f1(x))
    x = np.array(x)
    return 1 / (1+np.exp(-(x**2)))

def df2f1(x):
    x = np.array(x)
    return (2*x*np.exp(-x**2)) / (1+np.exp(-x**2))**2

```

```python
x = np.arange(-3,3,0.1)
y = f2f1(x)
dy = df2f1(x)

#print(x.size, y.size)
plt.plot(x,y)
plt.plot(x,dy)
plt.grid(True)
```

```python
x,y,a,b,c = sy.symbols('x y a b c ')
f2f1 = sy.diff(1 / (1+sy.exp(-(x**2))), x)

f = sy.lambdify(x, f2f1)
f(0)
```

```python
X = np.arange(-3,3,0.1)
dy = f(X)
plt.plot(X,dy)
plt.grid(True)
```
